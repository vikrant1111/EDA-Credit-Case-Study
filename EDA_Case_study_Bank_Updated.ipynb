```python
#We can create a new colum AMT_ANNUITY_CATEGORICAL by dividing the values of AMT_ANNUITY columns to get a better insight

#Function that returns "Low" for annuity less than 20000 , returns Medium for annuity between 20000 and 50000, returns High for annuity between 50000 and 100000 and returns "Very high" for all the values. 
def annuity_amt(x):
    if x <= 20000:
        return "Low"
    elif x > 20000 and x <=50000:
        return "Medium"
    elif x > 50000 and x <=100000:
        return "High"
    else:
        return "Very High"
    
app_data["AMT_ANNUITY_CATEGORICAL"] = app_data["AMT_ANNUITY"].apply(annuity_amt)
app_data["AMT_ANNUITY_CATEGORICAL"].head()
```

**Now we will create three subplots for each categorical column**

*1. Subplot : Here we will be plotting a **pie chart** for that categorical column in terms of percentage.*

*2. Subplot : Here we will be plotting a **bar chart** for the value counts of each category in that categorical column with respect to clients with difficulties and without difficulties.*

*3. Subplot : Here we will be plotting a **bar chart** for percentage of each category with respect to clients with difficulties and without difficulties.*

```python
#Creating a function to plot three subplots mentioned above

def plotting(train, train0, train1, column):
    
    train = train
    train_0 = train0
    train_1 = train1
    col = column
    
    fig = plt.figure(figsize=(10,10))
    
    ax1 = plt.subplot(221)
    train[col].value_counts().plot.pie(autopct = "%1.0f%%", ax=ax1)
    #plt.title('Plotting data for the column: '+ column)
    plt.title("Percentage of each category in "+ column, fontdict={"fontsize":12, "fontweight":5, "color":"blue"})
    

    
    ax2 = plt.subplot(222)
    sns.countplot(x= column, hue = 'TARGET', data = train, ax = ax2)
    plt.xticks(rotation=90)
    plt.title('Plotting data for target in terms of total count', fontdict={"fontsize":12, "fontweight":5, "color":"blue"})
    plt.xlabel(column, fontdict={"fontsize":10, "fontweight":5, "color":"brown"})
    plt.ylabel("Counts", fontdict={"fontsize":10, "fontweight":5, "color":"brown"})


    ax3 = plt.subplot(223)
    df = pd.DataFrame()
    df['0']= ((train_0[col].value_counts())/len(train_0))
    df['1']= ((train_1[col].value_counts())/len(train_1))
    df.plot.bar(ax=ax3)
    plt.title('Plotting data for target in terms of percentage', fontdict={"fontsize":12, "fontweight":5, "color":"blue"})
    plt.xlabel(column, fontdict={"fontsize":10, "fontweight":5, "color":"brown"})
    plt.ylabel("Percentage", fontdict={"fontsize":10, "fontweight":5, "color":"brown"})



    fig.tight_layout() # Or equivalently,  "plt.tight_layout()"

    plt.show()
```

```python
SRno = 1
for column in categorical:
    print(SRno,"Plotting", column)
    plotting(app_data, app_data_1, app_data_0, column)
    print("\n=============================================================================================\n")
    SRno+=1
```

```python
#Function that returns barplot for each categorical column from list categorical_col and column NAME_CONTRACT_STATUS from previous application dataset.

def bivariate_merge(columns):
    analysis=pd.crosstab(index=merge_data['NAME_CONTRACT_STATUS'],columns=merge_data[columns])

    plot1=analysis.plot(kind="bar", figsize=(10,8))
    plt.title('NAME_CONTRACT_STATUS Vs '+columns, fontdict={"fontsize":12, "fontweight":5, "color":"blue"})
    plt.xlabel('NAME_CONTRACT_STATUS', fontdict={"fontsize":10, "fontweight":5, "color":"brown"})
    plt.ylabel(columns+' (counts)', fontdict={"fontsize":10, "fontweight":5, "color":"brown"})
    
    return plot1

for column in categorical_col:
    bivariate_merge(column)
```
